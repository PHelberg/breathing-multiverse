This notebook simulates predictions from the **Breathing Multiverse Theory**, a cosmological model where universes undergo expansion and contraction cycles. A kind of "breathing" motion.

###  Core Concepts
- Universes behave like interacting bubbles in a shared quantum field.
- Breathing cycles produce:
  - **Gravitational waves** (low-frequency background)
  - **Dark matter** from inter-bubble friction
  - **CMB anomalies** from quantum-scale interference

###  Modeling Goals
- Simulate `a(t)` as a breathing exponential scale factor.
- Compare gravitational wave strain against known models (like NANOGrav).
- Visualize dark matter and CMB distortions as outputs of this dynamic process.



```python
import numpy as np
import matplotlib.pyplot as plt

# Time from 0 to 20 billion years (in Gyr), 1000 points
t = np.linspace(0, 20, 1000)

# Simulated breathing scale factor a(t)
# Think of this as a rough oscillation around an expanding baseline
a = (1 + 0.3 * np.sin(2 * np.pi * t / 5)) * np.exp(0.03 * t)

# Plot the result
plt.figure(figsize=(10, 5))
plt.plot(t, a, label='Breathing Scale Factor a(t)', color='darkblue')
plt.xlabel('Time (Gyr)')
plt.ylabel('Scale Factor a(t)')
plt.title('Simulated Breathing Universe Expansion')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Gravitational Wave Background Prediction

This plot compares the predicted gravitational wave background from the Breathing Multiverse Theory  
to a standard reference (similar to what NANOGrav expects).  
A sinusoidal breathing pattern modulates the strain at nanohertzâ€“microhertz frequencies.


```python
# Simulated gravitational wave background (GW) from breathing cycles

# Define frequency range (nanohertz to microhertz scale)
f = np.logspace(-9, -6, 500)  # Hz

# Your predicted GW power spectrum: modeled as sinusoidal modulation with decay
gw_breathe = 1e-14 * (1 + 0.3 * np.sin(2 * np.pi * np.log10(f * 1e9))) * (f / 1e-8)**(-2/3)

# Simulated reference curve (based on expected NANOGrav background)
gw_ref = 1e-14 * (f / 1e-8)**(-2/3)

# Plot the two curves
plt.figure(figsize=(10, 5))
plt.loglog(f, gw_ref, '--', label='Reference GWB (NANOGrav-style)', color='gray')
plt.loglog(f, gw_breathe, label='Breathing Theory Prediction', color='crimson')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Strain Power')
plt.title('Gravitational Wave Background: Breathing vs Reference')
plt.grid(True, which='both')
plt.legend()
plt.tight_layout()
plt.show()
```

## Dark Matter from Inter-Universal Friction

This model simulates dark matter as a byproduct of **friction between interacting universes**.  
As the scale factor oscillates (â€œbreathesâ€), energy is lost to friction,  
which accumulates as what we perceive as **dark matter**.

- The breathing causes frequent directional shifts.
- Energy is lost to inter-universal resistance.
- That energy builds up over time as an effective mass field.



```python
# Simulate dark matter density growth from inter-universal friction

time = np.linspace(0, 20, 1000)  # Gyr
friction_strength = 0.015  # Adjustable constant to simulate energy transfer rate

# Basic breathing function for expansion
scale_factor = (1 + 0.3 * np.sin(2 * np.pi * time / 5)) * np.exp(0.03 * time)

# Simulated dark matter density: accumulates over time based on energy lost to friction
dark_matter_density = np.cumsum(friction_strength * np.abs(np.gradient(scale_factor)))  # arbitrary units

plt.figure(figsize=(10, 5))
plt.plot(time, dark_matter_density, label='Simulated Dark Matter Density', color='purple')
plt.xlabel('Time (Gyr)')
plt.ylabel('Dark Matter Density (a.u.)')
plt.title('Dark Matter Density from Inter-Bubble Friction')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Simulated CMB Distortion from Breathing

This section simulates how **CMB patterns** might be distorted  
by the breathing behavior of the multiverse.  

The idea:  
As universes expand and contract, their motion affects the background radiation left over from the Big Bang.  
This motion could leave **interference patterns** or **cold spots** in the CMB. Like the ones seen by Planck.

The colors in this simulation represent:
- Oscillation phases across spacetime
- Possible hot/cold regions caused by energy exchange


```python
# Simulated CMB distortion caused by breathing universe dynamics

# Create a 2D sky grid
grid_size = 200
x = np.linspace(-1, 1, grid_size)
y = np.linspace(-1, 1, grid_size)
X, Y = np.meshgrid(x, y)

# Polar coordinates: radius and angle
r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

# Simulate temperature fluctuations due to breathing cycles
temperature_map = (
    1
    + 0.05 * np.sin(10 * r * np.pi)                 # radial ripples
    + 0.08 * np.exp(-5 * r**2) * np.cos(5 * theta)  # breathing anisotropy
)

# Plot the result
plt.figure(figsize=(6, 6))
plt.imshow(temperature_map, cmap='coolwarm', extent=[-1, 1, -1, 1])
plt.title('Simulated CMB Distortion from Breathing Cycles')
plt.colorbar(label='Î”T (Normalized Temperature Units)')
plt.xlabel('Sky X Direction')
plt.ylabel('Sky Y Direction')
plt.tight_layout()
plt.show()
```

```python
# Interactive plot of a breathing universe scale factor over time

def breathing_scale(t, A=0.3, f=0.2, H=0.03):
    """Breathing scale factor: exponential expansion with sinusoidal modulation"""
    return (1 + A * np.sin(2 * np.pi * f * t)) * np.exp(H * t)

# Set up time array
t = np.linspace(0, 20, 1000)  # Gyr

# Create interactive plot
@interact(time=(0, 20, 0.1))
def plot_breathing(time=10):
    a = breathing_scale(t)

    plt.figure(figsize=(8, 4))
    plt.plot(t, a, label='Breathing Scale Factor')
    plt.axvline(x=time, color='red', linestyle='--', label=f'Time = {time:.1f} Gyr')
    plt.scatter(time, breathing_scale(time), color='red')

    plt.xlabel('Time (Gyr)')
    plt.ylabel('Scale Factor a(t)')
    plt.title('Interactive Breathing Universe Simulation')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

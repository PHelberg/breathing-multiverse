# The Breathing Multiverse Theory

This notebook simulates predictions from the **Breathing Multiverse Theory**, a cosmological model where universes undergo expansion and contraction cycles. A ‚Äúbreathing‚Äù motion.

### Core Ideas:
- Universes are like interacting bubbles in a shared quantum field.
- These cycles produce:
  - **Gravitational waves** (especially low-frequency)
  - **Dark matter** as friction between universes
  - **CMB anomalies** due to energy interference patterns

### Mathematical Framework:
- Scale factor `a(t)` is modeled as a breathing sine wave on top of exponential expansion.
- Gravitational wave signal is overlaid against NANOGrav‚Äôs expected background.
- Future versions will simulate:
  - Dark matter density evolution
  - CMB distortion maps
  - Parameter tuning via sliders

```python
import numpy as np
import matplotlib.pyplot as plt

# Time from 0 to 20 billion years (in Gyr), 1000 points
t = np.linspace(0, 20, 1000)

# Simulated breathing scale factor a(t)
# Think of this as a rough oscillation around an expanding baseline
a = (1 + 0.3 * np.sin(2 * np.pi * t / 5)) * np.exp(0.03 * t)

# Plot the result
plt.figure(figsize=(10, 5))
plt.plot(t, a, label='Breathing Scale Factor a(t)', color='darkblue')
plt.xlabel('Time (Gyr)')
plt.ylabel('Scale Factor a(t)')
plt.title('Simulated Breathing Universe Expansion')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Simulated gravitational wave background (GW) from breathing cycles

# Define frequency range (nanohertz to microhertz scale)
f = np.logspace(-9, -6, 500)  # Hz

# Your predicted GW power spectrum: modeled as sinusoidal modulation with decay
gw_breathe = 1e-14 * (1 + 0.3 * np.sin(2 * np.pi * np.log10(f * 1e9))) * (f / 1e-8)**(-2/3)

# Simulated reference curve (based on expected NANOGrav background)
gw_ref = 1e-14 * (f / 1e-8)**(-2/3)

# Plot the two curves
plt.figure(figsize=(10, 5))
plt.loglog(f, gw_ref, '--', label='Reference GWB (NANOGrav-style)', color='gray')
plt.loglog(f, gw_breathe, label='Breathing Theory Prediction', color='crimson')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Strain Power')
plt.title('Gravitational Wave Background: Breathing vs Reference')
plt.grid(True, which='both')
plt.legend()
plt.tight_layout()
plt.show()
```

## Dark Matter from Breathing Friction

This model simulates dark matter as a byproduct of **friction between interacting bubbleverses**.  
As the scale factor oscillates ("breathes"), energy is lost to inter-bubble resistance,  
which accumulates as what we perceive as **dark matter**.

- The breathing pattern creates frequent directional shifts.
- The greater the oscillation, the more energy lost to friction.
- The density grows over time, stabilizing as breathing amplitude decreases.

This is an original, testable explanation for dark matter that *emerges from multiversal mechanics*.
## Dark Matter from Breathing Friction

```python
# Simulate dark matter density growth from inter-universal friction

time = np.linspace(0, 20, 1000)  # Gyr
friction_strength = 0.015  # Adjustable constant to simulate energy transfer rate

# Basic breathing function for expansion
scale_factor = (1 + 0.3 * np.sin(2 * np.pi * time / 5)) * np.exp(0.03 * time)

# Simulated dark matter density: accumulates over time based on energy lost to friction
dark_matter_density = np.cumsum(friction_strength * np.abs(np.gradient(scale_factor)))  # arbitrary units

plt.figure(figsize=(10, 5))
plt.plot(time, dark_matter_density, label='Simulated Dark Matter Density', color='purple')
plt.xlabel('Time (Gyr)')
plt.ylabel('Dark Matter Density (a.u.)')
plt.title('Dark Matter Density from Inter-Bubble Friction')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Simulated CMB Distortion from Breathing

This visualization shows how breathing cycles could leave **observable imprints** in the Cosmic Microwave Background (CMB).

- Radial sine waves represent energy interference from scale factor oscillation.
- Asymmetry is created by directional "pulses" from nearby universes affecting spacetime curvature.
- Central zones show compressed regions ‚Äî analogous to cold spots or anomalies in Planck data.

This toy model demonstrates how breathing behavior might show up as **patterned distortions** in the large-scale structure of the CMB.

```python
# Simulated CMB distortion caused by breathing universe dynamics

# Create a 2D grid of "sky" pixels
grid_size = 200
x = np.linspace(-1, 1, grid_size)
y = np.linspace(-1, 1, grid_size)
X, Y = np.meshgrid(x, y)

# Simulate breathing interference ‚Äî ripples radiating from center
r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

# Central breathing hotspot effect + radial sine wave
temperature_map = (
    1 + 0.05 * np.sin(10 * r * np.pi) +          # radial wave
    0.08 * np.exp(-5 * r**2) * np.cos(5 * theta) # breathing anisotropy
)

# Plot heatmap
plt.figure(figsize=(6, 6))
plt.imshow(temperature_map, cmap='coolwarm', extent=[-1, 1, -1, 1])
plt.title('Simulated CMB Distortion from Breathing Cycles')
plt.colorbar(label='ŒîT (normalized units)')
plt.xlabel('X (sky direction)')
plt.ylabel('Y (sky direction)')
plt.tight_layout()
plt.show()
```

## Interactive Breathing Universe Simulator

Use the sliders below to explore how the universe might expand and contract over time in the **Breathing Multiverse Theory**.

- **Frequency** controls how often the universe breathes (in Gyr‚Åª¬π)
- **Amplitude** controls how strong each breathing cycle is

This interactive model lets you see how the scale factor `a(t)` changes based on the parameters of your theory.  
Try increasing the amplitude and you'll see how extreme the contractions become. Lower the frequency to simulate longer, slower breathing cycles.

This kind of tuning could affect gravitational wave signals, dark matter production, or even CMB structure. All of which are explored in earlier cells.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def plot_breathing_universe(frequency=5.0, amplitude=0.3):
    t = np.linspace(0, 20, 1000)
    a = (1 + amplitude * np.sin(2 * np.pi * frequency * t / 20)) * np.exp(0.03 * t)

    plt.figure(figsize=(10, 5))
    plt.plot(t, a, label=f'Breathing a(t)\nFreq={frequency} Gyr‚Åª¬π, Amp={amplitude}')
    plt.xlabel('Time (Gyr)')
    plt.ylabel('Scale Factor a(t)')
    plt.title('Interactive Breathing Universe Model')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Create sliders
frequency_slider = widgets.FloatSlider(value=5.0, min=0.1, max=10.0, step=0.1, description='Frequency')
amplitude_slider = widgets.FloatSlider(value=0.3, min=0.0, max=1.0, step=0.05, description='Amplitude')

ui = widgets.VBox([frequency_slider, amplitude_slider])
out = widgets.interactive_output(plot_breathing_universe, {'frequency': frequency_slider, 'amplitude': amplitude_slider})

display(ui, out)
```
